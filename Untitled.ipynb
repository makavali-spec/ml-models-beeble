```python
#Detecting outliers
from sklearn.cluster import DBSCAN
import random
random.seed(1)
random_data = df

outlier_detection = DBSCAN(min_samples = 2, eps = 3)
clusters = outlier_detection.fit_predict(random_data)
list(clusters).count(-1)
```

```python
#there are outliers in the dataset
```

```python
#All values of the dataset must be represented as numerical variables.
#There's no column with a good use case of catergorical variable.
```

```python
#numerical transformation
df['RAD'] = pd.to_numeric(df['RAD'], errors='coerce')
```

```python
#Binning
df["AGE_BINNED"] = pd.cut(df['AGE'], bins=[2.9,12,20,100], labels=['New','Intermediate','Old'])
df["AGE_BINNED"]
```
